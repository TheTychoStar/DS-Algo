## Deque  - "deck"

```python
queue.extendleft('one')
queue                  
```

## Tree

```python
# Definition for a binary tree node.
class TreeNode(object):
    def __init__(self, x):
        self.val = x
        self.left = None
        self.right = None
```

```python
tr = TreeNode('A')
tr
```

```python
tr.left = TreeNode('B')
tr.right = TreeNode('C')
tr
```

```python
tr.val , tr.left.val, tr.right.val 
```

```python
def DFS ( tr):
    if tr == None:
        return
    print(tr.val)
    DFS(tr.left)
    DFS(tr.right)
    
```

```python
DFS(tr)
```
